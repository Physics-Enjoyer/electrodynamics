## Vector and Scaler Quantities

A **scaler** is a mathematical object that has only a magnitude. In other words it is just a _number_

Examples of scalers

- Temperature
- Energy
- Quantity of something
- Speed

A **vector** on the other hand is a mathematical object that has a magnitude _and_ a direction.

Examples of vectors

- Velocity (not speed)
- Momentum
- Angular Momentum
- Acceleration
- Position

There are many more examples of both vectors and scalers but thats enough for now.

A **unit vector** is a vector with a magnitude of $1$. Unit vectors are dimensionless and are calculated by dividing the vector by the magnitude of the vector.

$$\hat{A} = \frac{\vec{A}}{|\vec{A}|}$$


## Vector Addition

Adding vectors can be accomplished a few different ways. Graphical or analytical.

### Graphical addition

Given two vectors $\vec{A}$ and $\vec{B}$ shown below:


```python
A = np.array([3, 4])
B = np.array([7, 5])
AplusB = A + B

fig, ax = plt.subplots()
ax.arrow(0, 0, A[0], A[1], head_width=0.1)
ax.annotate("$\\vec{A}$", (1.25, 2))
ax.arrow(0, 0, B[0], B[1], head_width=0.1)
ax.annotate("$\\vec{B}$", (2.2, 1.75))
ax.set_aspect("equal")
ax.set_title("Two vectors $\\vec{A}$ and $\\vec{B}$")
plt.show()
```

to add the vectors together put the tail of $\vec{B}$ on the head of $\vec{A}$ and draw a line from the tail of $\vec{A}$ to the head of $\vec{B}$ as shown below.


```python
fig, ax = plt.subplots()
ax.arrow(0, 0, A[0], A[1], head_width=0.1)
ax.annotate("$\\vec{A}$", (1.25, 2))
ax.arrow(A[0], A[1], B[0], B[1], head_width=0.1)
ax.annotate("$\\vec{B}$", (6,6.5))
ax.arrow(0, 0, AplusB[0], AplusB[1], head_width=0.1)
ax.annotate("$\\vec{A} + \\vec{B}$", (4, 2.75))
ax.set_aspect("equal")
ax.set_title("Addition of $\\vec{A}$ and $\\vec{B}$")
plt.show()
```

You can add more than 2 vectors together at a time by _chaining_ Them together.


```python
C = np.array([2, -4])
ABC = A + B + C

fig, ax = plt.subplots()
ax.arrow(0, 0, A[0], A[1], head_width=0.1)
ax.annotate("$\\vec{A}$", (1.25, 2))
ax.arrow(A[0], A[1], B[0], B[1], head_width=0.1)
ax.annotate("$\\vec{B}$", (6,6.5))
ax.arrow(AplusB[0], AplusB[1], C[0], C[1], head_width=0.1)
ax.annotate("$\\vec{C}$", (11.25, 6.75))
ax.arrow(0, 0, ABC[0], ABC[1], head_width= 0.1)
ax.annotate("$\\vec{A}+\\vec{B}+\\vec{C}$", (6.5, 2))
ax.set_aspect("equal")
ax.set_title("Addition of 3 vectors $\\vec{A}$, $\\vec{B}$, and $\\vec{C}$")
plt.show()
```

Subtracting one vector from another $\vec{A}-\vec{B}$ is just addition of the additive inverse of one of the second vector. $\vec{A} + (-1)\vec{B}$ which reverses the direction of the vector.


```python
negB = -B
AminusB = A + negB

fig, ax = plt.subplots()
ax.arrow(0, 0, B[0], B[1], head_width=0.1)
ax.annotate("$\\vec{B}$", (2.25,1.8))
ax.arrow(B[0]+1, B[1], negB[0], negB[1], head_width=0.1)
ax.annotate("$-\\vec{B}$", (3, 1.25))
ax.set_aspect("equal")
ax.set_title("$\\vec{B}$ and $-\\vec{B}$")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.arrow(0, 0, A[0], A[1], head_width=0.1)
ax.annotate("$\\vec{A}$", (1.75, 1.5))
ax.arrow(A[0], A[1], negB[0], negB[1], head_width=0.1)
ax.annotate("$-\\vec{B}$", (0, 1.75))
ax.arrow(0, 0, AminusB[0], AminusB[1], head_width=0.1)
ax.annotate("$\\vec{A}-\\vec{B}$", (-1.5, -0.75))
ax.set_aspect("equal")
ax.set_title("Subtraction of $\\vec{B}$ from $\\vec{A}$")
plt.show()
```

### Properties of vector addition

Vector addition is

- Associative
  - $\vec{A} + \vec{B} + \vec{C} = (\vec{A}+\vec{B})+\vec{C} = \vec{A}+(\vec{B}+\vec{C})$
- Commutative
  - $\vec{A}+\vec{B} = \vec{B} + \vec{A}$


## Scaler Multiplication

A vector can be multiplied by a scaler

$k\vec{A} = k(A\hat{A}) = (kA)\hat{A}$

Where $k$ is some scaler quantity, $A$ is the magnitude of the vector $\vec{A}$ and $\hat{A}$ is the unit vector pointing in the direction of $\vec{A}$. Note that $k$ only changes the magnitude and not the direction except in the case of a negative scaler than it reverses the direction.


## Dot product of two vectors

Two vectors can be multiplied together via the **Dot Product**.

The dot product is defined as $\vec{A}\cdot\vec{B} = AB\cos(\theta)$ where $\theta$ is the angle between the two vectors and $AB$ is $|\vec{A}||\vec{B}|$ the product of the magnitude of each vector.


```python
A = np.array([5, 0])
B = np.array([2, 2])
r = 0.5
theta = np.linspace(0, np.arctan(1))
x = r*np.cos(theta)
y = r*np.sin(theta)
fig, ax = plt.subplots()
ax.arrow(0, 0, A[0], A[1], head_width=0.1)
ax.annotate("$\\vec{A}$", (1,1.2))
ax.arrow(0, 0, B[0], B[1], head_width=0.1)
ax.annotate("$\\vec{B}$", (2.5, 0.1))
ax.arrow(0, 0, 2, 0, head_width=0.1, color="red")
ax.annotate("$A\\cos(\\theta)$", (1, 0.1))
ax.plot(x,y)
ax.annotate("$\\theta$", (0.6, 0.25))
ax.vlines(2, 0, 2, linestyles=":")
ax.set_aspect("equal")
ax.set_title("Dot Prodcut of 2 vectors")
plt.show()
```

### Properties of the dot product

- The dot product of two perpendicular vectors is always zero
- If the angle between the two vectors is greater than $90^{\circ}$ then the dot product is negative
- Dot products are commutative
  - $\vec{A}\cdot\vec{B} = \vec{B}\cdot\vec{A}$
- Dot products are distributive
  - $\vec{A}\cdot(\vec{B} + \vec{C}) = \vec{A}\cdot\vec{B} + \vec{A}\cdot\vec{C}$


## Cross product of two vectors

The cross product is defined as

$$\vec{A}\times\vec{B} = AB\sin(\theta)\hat{n}$$

here $AB$ is the same as the dot product, i.e. the product of the magnitudes of each of the vectors but instead of multiplying by $\cos{\theta}$ its multiplied by $\sin(\theta)$. The $\hat{n}$ is the unit vector that is perpendicular to the plane made by the two vectors and can be found using the _right hand rule_ that we learned about in earlier physics classes.

The magnitude of the cross product is the area of the parallelogram formed by the two vectors as shown below.


```python
A = np.array([1, 2])
B = np.array([4, 0])

fig, ax = plt.subplots()
ax.arrow(0, 0, A[0], A[1], head_width=0.1, color="red")
ax.arrow(0, 0, B[0], B[1], head_width=0.1, color="blue")
ax.arrow(A[0], A[1], B[0], B[1], head_width=0.1, color="blue", ls="--")
ax.arrow(B[0], B[1], A[0], A[1], head_width=0.1, color="red", ls="--")
ax.annotate("$\\vec{A}$", (0.25, 1))
ax.annotate("$\\vec{B}$", (2, 0.1))
ax.set_aspect("equal")
ax.set_title("Cross Product of $\\vec{A}$ and $\\vec{B}$")
plt.show()
```

### Properties of the cross product

- The cross product is **not** commutative and is in fact _anti_-commutative
  - $\vec{A}\times\vec{B} \neq \vec{B}\times\vec{A}$
  - $\vec{A}\times\vec{B} = -\vec{B}\times\vec{A}$
- The cross product is distributive
  - $\vec{A}\times(\vec{B} + \vec{C}) = \vec{A}\times\vec{B} + \vec{A}\times\vec{C}$


## Vector componenets

Any vector can be written as a sum of other vectors. Put in another way any vector can be written as a linear combination of other vectors. For example given a vector $\vec{D} = (3, 2)$ this can be written as a linear combination of $\vec{A} = (0, 2)$ and $\vec{B} = (3, 0)$. However it can also be written as a combination of the vectors $\vec{A} = (-1, 2)$ and $\vec{B} = (4, 0)$ or as $\vec{A} = (3, 2)$ and $\vec{B} = (1, -1)$

These three examples are shown below:


```python
A1 = np.array([0, 2])
B1 = np.array([3, 0])
D1 = A1+B1

A2 = np.array([-1, 2])
B2 = np.array([4, 0])
D2 = A2+B2

A3 = np.array([4, 1])
B3 = np.array([-1, 1])
D3 = A3+B3

fig, ax = plt.subplots(1, 3, figsize=(9,3))
for a in ax.flatten():
    a.set_aspect("equal")
    a.set_xlim(-1.5, 4.5)
    a.set_ylim(-1.5, 4.5)
ax[0].arrow(0, 0, A1[0], A1[1], head_width=0.1, color="red")
ax[0].arrow(0, 0, B1[0], B1[1], head_width=0.1, color="blue")
ax[0].arrow(0, 0, D1[0], D1[1], head_width=0.1)
ax[0].arrow(A1[0], A1[1], B1[0], B1[1], head_width=0.1, color="blue", ls="--")
ax[0].arrow(B1[0], B1[1], A1[0], A1[1], head_width=0.1, color="red", ls="--")
ax[1].arrow(0, 0, A2[0], A2[1], head_width=0.1, color="red")
ax[1].arrow(0, 0, B2[0], B2[1], head_width=0.1, color="blue")
ax[1].arrow(0, 0, D2[0], D2[1], head_width=0.1)
ax[1].arrow(A2[0], A2[1], B2[0], B2[1], head_width=0.1, color="blue", ls="--")
ax[1].arrow(B2[0], B2[1], A2[0], A2[1], head_width=0.1, color="red", ls="--")
ax[2].arrow(0, 0, A3[0], A3[1], head_width=0.1, color="red")
ax[2].arrow(0, 0, B3[0], B3[1], head_width=0.1, color="blue")
ax[2].arrow(0, 0, D3[0], D3[1], head_width=0.1)
ax[2].arrow(A3[0], A3[1], B3[0], B3[1], head_width=0.1, color="blue", ls="--")
ax[2].arrow(B3[0], B3[1], A3[0], A3[1], head_width=0.1, color="red", ls="--")
fig.suptitle("Vectors as a linear combination of other vectors")
fig.tight_layout()
plt.show()
```

Note that this is also valid for in 3D space as well.


The **magnitude** of a vector is given by the generalized Pythagorean Theorem.
$|\vec{D}| = \sqrt{A^2+B^2}$ or in 3D space $|\vec{D}| = \sqrt{A^2+B^2+C^2}$

Vectors can be written in component form such as $\vec{A} = A_x\hat{x}+A_y\hat{y}+A_z\hat{z}$ where $\hat{x}$, $\hat{y}$, and $\hat{z}$ are the unit vectors along the $x$, $y$, and $z$ axes.


```python
import numpy as np
import matplotlib.pyplot as plt

x_hat = np.array([1, 0, 0])
y_hat = np.array([0, 1, 0])
z_hat = np.array([0, 0, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scale_factor = 1

ax.quiver(0, 0, 0, *x_hat, color='r', label='$\\hat{x}$', length=scale_factor, arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, *y_hat, color='g', label='$\\hat{y}$', length=scale_factor, arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, *z_hat, color='b', label='$\\hat{z}$', length=scale_factor, arrow_length_ratio=0.1)

ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.5, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(azim=45)
ax.legend()
ax.set_title("Unit Vectors in 3D Space")
fig.tight_layout()
plt.show()
```

With components many of the operations described above become much easier.

Vector addition becomes:

$\vec{A}+\vec{B} = (A_x+B_x)\hat{x} + (A_y+B_y)\hat{y} + (A_z+B_z)\hat{z}$

The dot product becomes

$\vec{A}\cdot\vec{B} = A_xB_x + A_yB_y + A_zB_z$

This can be explained by:

$$
\vec{A}\cdot\vec{B} = (A_x\hat{x} + A_y\hat{y} + A_z\hat{z})(B_x\hat{x} + B_y\hat{y} + B_z\hat{z})
$$

$$
= (A_xB_x)\hat{x}\cdot\hat{x} + (A_xB_y)\hat{x}\cdot\hat{y} +(A_xB_z)\hat{x}\cdot\hat{y} \\
+ (A_yB_x)\hat{y}\cdot\hat{x} + (A_yB_y)\hat{y}\cdot\hat{y} + (A_yB_z)\hat{y}\cdot\hat{z} \\
+ (A_zB_x)\hat{z}\cdot\hat{x} + (A_zB_y)\hat{z}\cdot\hat{y} + (A_zB_z)\hat{z}\cdot\hat{z}
$$

But we know that $\hat{x}\cdot\hat{y} = \hat{x}\cdot\hat{z} = \hat{y}\cdot\hat{z} = 0$ so the only terms that remain are the terms with $\hat{x}\cdot\hat{x} = \hat{y}\cdot\hat{y} = \hat{z}\cdot\hat{z} = 1$ and the final result is:

$\vec{A}\cdot\vec{B} = A_xB_x + A_yB_y + A_zB_z$

We can express this much more compactly using summation notation.

$$
\vec{A}\cdot\vec{B} = \sum_{i=1}^3A_iB_i
$$
