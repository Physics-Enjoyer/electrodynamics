## Section 1.1.1 Vector Operations

**Vectors** are mathematical objects that follow certain properties. In physics vectors are generally described as something that has both a _magnitude_ and a _direction_.

Examples of vectors are:

- Displacement (not distance)
- Velocity
- Acceleration
- Force
- Momentum

Objects that are just a numerical value and do not have a direction associated with them are called **scalers**.

Examples of scalers are:

- Mass
- Charge
- Density
- Temperature

Vectors can be notated by either using **boldface** letters (e.g. $\bf{r}$, $\bf{v}$, $\bf{a}$, $\bf{F}$, $\bf{p}$) or placing an arrow over the letter (e.g. $\vec{r}$, $\vec{v}$, $\vec{a}$, $\vec{F}$, $\vec{p}$).

The magnitude of a vector can be notated either by putting vertical bars around it such as |**a**| or simply using non-boldface letters $a$.

A vector that has the same magnitude of $\bf{a}$ but points in the opposite direction of $\bf{a}$ is $-\bf{a}$

Examples of vectors are below:


```python
# Plot 2 vectors
x0 = 0
y0 = 0
x1 = 1
y1 = 0
dx = 1
dy = 1

fig, ax = plt.subplots(1, 2)
for a in ax:
    a.arrow(x0, y0, dx, dy, head_width=0.025)
    a.set_aspect("equal")
    a.axis(False)
ax[0].set_title("The vector $\\bf{a}$ in two spots")
ax[0].arrow(x1, y1, dx, dy, head_width=0.025)
ax[1].arrow(x1 + dx, y1 + dx, -dx, -dy, head_width=0.025)
ax[1].set_title("The vectors $\\bf{a}$ and $\\bf{-a}$")
fig.tight_layout()
plt.show()
```

For our purposes we define four main vector operations. The four operations are:

- Addition
- Three types of multiplication
  - Scaler multiplication
  - The **dot product** (also known as the **scaler product**)
  - the **cross product** (also known as the **vector product**)

Addition and scaler multiplication are both relatively simple.


### Vector addition

Two vectors can be added together by adding the _components_ of one vector to the _components_ of another vector. But what is a vector _component_? Well if you have made it to these notes you _should_ already know what a component is. It is the magnitude of the vector _in a particular direction_.

There are two methods for adding vectors. Graphical addition of vectors and analytical addition of vectors. I will first discuss graphical addition then move on to analytical addition which will likely be used for the remainder of the course.


#### Graphical Addition of Vectors

Graphical addition of vectors is fairly simple. To add two vectors $\bf{A}$ and $\bf{B}$ place the tail of $\bf{B}$ on the tip of $\bf{A}$


```python
A = np.array([3, 4])
B = np.array([2, 1])

fig, ax = plt.subplots(figsize=(3, 4))
ax.arrow(0, 0, A[0], A[1], head_width=0.025)
ax.annotate("$\\bf{A}$", (1, 1.5))
ax.arrow(0, 0, B[0], B[1], head_width=0.025)
ax.annotate("$\\bf{B}$", (1.25, 0.75))
ax.set_title("Two vectors $\\bf{A}$ and $\\bf{B}$")
ax.set_aspect("equal")
plt.show()
```

To add the two vectors we move the tail of **B** to the head of **A**


```python
C = A + B

fig, ax = plt.subplots(figsize=(5, 5))
ax.arrow(0, 0, A[0], A[1], head_width=0.05)
ax.annotate("$\\bf{A}$", (2, 3))
ax.arrow(A[0], A[1], B[0], B[1], head_width=0.05)
ax.annotate("$\\bf{B}$", (4.25, 4.75))
ax.arrow(0, 0, C[0], C[1], head_width=0.05)
ax.annotate("$\\bf{C}$", (2.5, 2.25))
ax.set_aspect("equal")
plt.show()
```

#### Analytical addition of vectors

If you look in the code block above this is what I actually already did in order to plot the _resultant_ vector **C**. Analytical addition is simply _element wise_ addition. Given two vectors $\bf{A} = (3, 4)$ and $\bf{B} = (2, 1)$ the resultant vector $\bf{C} = (3+2, 4+1)$

In more general terms if $\bf{A} = (a_1, a_2, a_3, ..., a_n)$ and $\bf{B} = (b_1, b_2, b_3, ..., b_n)$ then $\bf{A}+\bf{B} = \bf{C} = (a_1+b_1, a_2+b_2, a_3+b_3, ..., a_n+b_n)$

This can be simplified with some notation shortcuts.

$$
\bf{C}_\mu = \bf{A}_\mu + \bf{B}_\mu
$$

Where $\mu$ is the _index_ of the vector.


#### Properties of vector addition

Vector addition has all the properties you would expect. It is commutative and associative. Meaning the order you add vectors doesn't matter no matter how many vectors you have.

$$
\bf{A} + \bf{B} = \bf{B} + \bf{A}
$$

$$
\bf{A} + (\bf{B} + \bf{C}) = (\bf{A} + \bf{B}) + \bf{C} = (\bf{A} + \bf{C}) + \bf{B}
$$


#### Vector subtraction

There's one more thing to mention about vector addition and thats vector subtraction. Vector subtraction is defined by the addition of the additive inverse. That means that $\bf{A} - \bf{B}$ really means $\bf{A} + (-\bf{B})$ and $-\bf{B} = -1\cdot\bf{B}$ or $-1$ times each element of $\bf{B}$ which is just multiplication by a scaler which is covered next.


### Vector Multiplication

There are 3 ways to multiply vectors

- Multiplication by a scaler
- Dot Product
- Cross Product

Multiplication by a scaler is multiplying a vector by a scaler or a number. It _scales_ the vector up or down. The other two vector multiplications, the dot product and the cross product, are ways of multiplying 2 vectors together.


#### Multiplication by a scaler

Given a vector $\bf{A}$ and a scaler $\alpha$ then scaler multiplication is simply the scaler $\alpha$ multiplied by each element of $\bf{A}$

$$
\bf{A} = (a_1, a_2, a_3, ..., a_n)
$$

then

$$
\alpha\bf{A} = (\alpha a_1, \alpha a_2, \alpha a_3, ..., \alpha a_n)
$$


#### Dot Product

To multiply 2 vectors together with the dot product is similar to vector addition only at the end we sum all the results together.

$$
\bf{A} \cdot \bf{B} = a_1b_1 + a_2b_2 + a_3b_3 + ..., +a_nb_n
$$

The dot product returns a scaler value. In other words a number. In summation notation this is:

$$
\bf{A} \cdot \bf{B} = \sum_{\mu}a_{\mu}b_{\mu}
$$

Another way of defining the dot product is by using geometry.

$$
\bf{A}\cdot\bf{B} = |A||B|\cos(\theta)
$$

Where $\theta$ is the angle between the two vectors.

##### Properties of the Dot Product

The dot product is commutative and distributive.

**Commutativity**

$$
\bf{A}\cdot\bf{B} = \bf{B}\cdot\bf{A}
$$

**Distributivity**

$$
\bf{A} \cdot (\bf{B} + \bf{C}) = \bf{A}\cdot\bf{B} + \bf{A}\cdot\bf{C}
$$


#### Cross Product

The cross product of two vectors is a way to multiply two vectors and get another vector. The definition of a cross product is

$$
\bf{A}\times\bf{B} = |\bf{A}||\bf{B}|\sin(\theta)\bf{\hat{n}}
$$

Where $\bf{\hat{n}}$ is a unit vector perpendicular to the plane made by $\bf{A}$ and $\bf{B}$

The direction of the cross product can be found by using the right hand rule.

##### Properties of the cross product

The cross product is distributive

$$
\bf{A}\times(\bf{B} + \bf{C}) = \bf{A}\times\bf{B} + \bf{A}\times\bf{C}
$$

but not commutative

$$
\bf{A}\times\bf{B} \neq \bf{B}\times\bf{A}
$$

In fact

$$\bf{A}\times\bf{B} = -(\bf{B}\times\bf{A})$$

Geometrically the cross product is the area of the parallelogram


```python
A = np.array([1, 3])
B = np.array([3, 0])

fig, ax = plt.subplots()
ax.arrow(0, 0, A[0], A[1], head_width=0.05, length_includes_head=True)
ax.arrow(0, 0, B[0], B[1], head_width=0.05, length_includes_head=True)
ax.arrow(A[0], A[1], B[0], B[1], head_width=0.05, length_includes_head=True, ls="--")
ax.arrow(B[0], B[1], A[0], A[1], head_width=0.05, length_includes_head=True, ls="--")
plt.show()
```

The cross product can also be defined via linear algebra as a solution to the determinant

$$
\bf{A}\times\bf{B} = \begin{vmatrix}
    \hat{\textbf{x}} & \hat{\textbf{y}} & \hat{\textbf{z}} \\
    a_x     & a_y     & a_z \\
    b_x     & b_y     & b_z \\
\end{vmatrix}

= \hat{x}\begin{vmatrix}
a_y & a_z \\
b_y & b_z
\end{vmatrix}
- \hat{y}\begin{vmatrix}
a_x & a_z \\
b_x & b_y
\end{vmatrix}
+ \hat{z}\begin{vmatrix}
a_x & a_y \\
b_x & b_y
\end{vmatrix}
$$

where $\hat{\textbf{x}}$, $\hat{\textbf{y}}$, and $\hat{\textbf{z}}$ are the unit vectors in the $x$, $y$, and $z$ direction.
